## Singular part of alpha calculation

```python
s = np.array(sp.symbols('s(1:3)(1:3)')).reshape((2,2))
xi = np.array(sp.symbols('xi(1:3)'))
x = np.array(sp.symbols('x(1:3)'))
c = sp.symbols('c')
a = sp.symbols('a')

r = np.array([[0, -1],[1, 0]], dtype=int)
```

```python
xi_x = s.T @ x
psi = np.sum((xi_x)*(xi_x))/2
alpha = c*sp.atan2(xi_x[1], xi_x[0])
grad_psi = np.array([sp.diff(psi, x[i]) for i in range(2)])
grad_alpha = np.array([sp.diff(alpha, x[i]) for i in range(2)])
cross = grad_psi[0]*grad_alpha[1] - grad_psi[1]*grad_alpha[0]
sp.simplify(cross)
```

```python
xi_x = s.T @ x
psi = (xi_x[1]*xi_x[1] - xi_x[0]*xi_x[0])/2
alpha = c*sp.log(xi_x[0] - xi_x[1])
grad_psi = np.array([sp.diff(psi, x[i]) for i in range(2)])
grad_alpha = np.array([sp.diff(alpha, x[i]) for i in range(2)])
cross = grad_psi[0]*grad_alpha[1] - grad_psi[1]*grad_alpha[0]
sp.simplify(cross)
```

## Tearing Mode Calculation

```python
from sympy.vector import CoordSys3D, Del, Laplacian, Curl
R = CoordSys3D('R')
```

```python
b0 = sp.Function('B_0y')(R.x) * R.j
k, gamma, t = sp.symbols('k gamma t')
eikonal = sp.I * k * R.y + gamma * t

def perturbed(label):
    return (sp.Function(f'{label}_1x')(R.x) * R.i + sp.Function(f'{label}_1y')(R.x) * R.j + sp.Function(f'{label}_1z')(R.x) * R.k) * sp.exp(eikonal)
v1 = perturbed('V')
b1 = perturbed('B')
```

```python
delop = Del()
eta, mu0 = sp.symbols('eta mu0')
R.i.dot((sp.diff(b1, t) - Curl(v1.cross(b0)).doit() - (eta/mu0) * Laplacian(b1).doit()).doit()).subs(sp.exp(eikonal), 1)
```

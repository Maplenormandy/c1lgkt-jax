```python
import sympy as sp
import matplotlib.pyplot as plt
import numpy as np
```

```python
x, y = sp.symbols('x y')


h11, h12, h21, h22 = sp.symbols('h11 h12 h21 h22')

a0, a1 = sp.symbols('a0 a1')

psi = x**2 - y**2
#psi = (h11 *x**2 + 2*h12*x*y + h22*y**2)/2
#alpha = sp.asinh(theta/psi)
alpha = a0 * sp.log(x+y) + a1 * sp.log(x - y)

grad_psi = sp.Array([sp.diff(psi, x), sp.diff(psi, y)])
grad_alpha = sp.Array([sp.diff(alpha, x), sp.diff(alpha, y)])

cross = grad_psi[0]*grad_alpha[1] - grad_psi[1]*grad_alpha[0]
sp.collect(sp.simplify(sp.expand(cross)), [x, y])
```

```python
s = np.array(sp.symbols('s(1:3)(1:3)')).reshape((2,2))
xi = np.array(sp.symbols('xi(1:3)'))
x = np.array(sp.symbols('x(1:3)'))
c = sp.symbols('c')
a = sp.symbols('a')

r = np.array([[0, -1],[1, 0]], dtype=int)
```

```python
xi_x = s.T @ x
psi = np.sum((xi_x)*(xi_x))/2
alpha = c*sp.atan2(xi_x[1], xi_x[0])
grad_psi = np.array([sp.diff(psi, x[i]) for i in range(2)])
grad_alpha = np.array([sp.diff(alpha, x[i]) for i in range(2)])
cross = grad_psi[0]*grad_alpha[1] - grad_psi[1]*grad_alpha[0]
sp.simplify(cross)
```

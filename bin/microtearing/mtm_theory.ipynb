# Theory Notebook for Microtearing Mode

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
```

```python
k = 0.5
xg = np.linspace(-5, 5, 256)
yg = np.linspace(-10, 10, 256)

xplot, yplot = np.meshgrid(xg, yg, sparse = True)

psitilde = np.exp(-k * np.abs(xplot)) * (1 + np.tanh(np.abs(xplot)) / k) * np.cos(k * yplot)
psi0 = np.log(np.cosh(xplot))

plt.contour(yg, xg, (psi0 + 0.1*psitilde).T, levels = 32)
plt.axis('equal')
```

```python
omega = np.exp(-k * np.abs(xplot)) * (1 + np.tanh(np.abs(xplot)) / k) / np.tanh(xplot) * np.sin(k * yplot)
plt.figure()
plt.contour(yg, xg, np.tanh(omega.T), levels=32)
```
